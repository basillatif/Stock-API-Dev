** Create a new dataframe called returns. This dataframe will contain the returns for each bank's stock. returns are typically defined by:**

### Moving Averages

** Please derive the moving averages for these stocks in the year 2016. Use .rolling() in pandas to get the rolling average calculation. ** 

```python
# your code here
pd.set_option('display.max_rows', None)
start = datetime.datetime(2016, 1, 1)
end = datetime.datetime(2016, 12, 31)
df = data.DataReader(['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC'],'yahoo', start, end)



close_px = df['Close']
bac_close_px = close_px['BAC']
bac_mavg = bac_close_px.rolling(window=30).mean()
c_close_px = close_px['C']
c_mavg = c_close_px.rolling(window=10).mean()
gs_close_px = close_px['GS']
gs_mavg = gs_close_px.rolling(window=10).mean()
jpm_close_px = close_px['JPM']
jpm_mavg = jpm_close_px.rolling(window=10).mean()
ms_close_px = close_px['MS']
ms_mavg = ms_close_px.rolling(window=10).mean()
wfc_close_px = close_px['WFC']
wfc_mavg = wfc_close_px.rolling(window=10).mean()
bac_mavg
```

** Plot the rolling 30 day average against the Close Price for Bank Of America's stock for the year 2016**

```python
# your code here
plt.plot(bac_mavg, bac_close_px, color='skyblue')
```

** Create a heatmap of the correlation between the stocks Close Price.**


```python
import seaborn as sns
#Create dataframe of close prices 
df_close = pd.concat([bac_close_px, c_close_px, gs_close_px, jpm_close_px, ms_close_px, wfc_close_px], axis=1)
# your code here
corr = df_close.corr()
fig, ax = plt.subplots(figsize=(10, 10))
#Generate Color Map
colormap = sns.diverging_palette(220, 10, as_cmap=True)
#Generate Heat Map, allow annotations and place floats in map
sns.heatmap(corr, cmap=colormap, annot=True, fmt=".2f")
#Apply xticks
plt.xticks(range(len(corr.columns)), corr.columns);
#Apply yticks
plt.yticks(range(len(corr.columns)), corr.columns)
#show plot
plt.show()
```
